# Extract features with overlap data

```python
import pandas as pd
import numpy as np

def fft_skew(fft_trans):
    colWise = []
    rowWise = []
    fft_mean, fft_std = fft_trans.mean(), fft_trans.std()
    for j in fft_trans:  # j th column among 22
        for x in fft_trans.iloc[:, j]:  # x is x value among 300

            rowWise.append(((x - fft_mean[j]) / fft_std[j]) ** 3)  # append the value for each rowwise computeation

        colWise.append(np.mean(rowWise))  # append the value for each feature
    return colWise


def fft_kurt(fft_trans):
    colWise = []
    rowWise = []
    fft_mean, fft_std = fft_trans.mean(), fft_trans.std()
    for j in fft_trans:  # j th column among 22
        for x in fft_trans.iloc[:, j]:  # x is x value among 300

            rowWise.append(((x - fft_mean[j]) / fft_std[j]) ** 4)  # append the value for each rowwise computeation

        colWise.append(np.mean(rowWise))  # append the value for each feature
    return colWise


def extraction(all_intervals ):
    all_features = []
    
    for interval in all_intervals:
        interval = pd.DataFrame(interval.data)
        interpolated = interval.interpolate(method='linear')
        interpolated = interpolated.fillna(method='backfill')
        fft_trans = np.abs(np.fft.fft(interpolated, axis=0))
        fft_trans = pd.DataFrame(fft_trans)
        feature = []
        # ==============time domain====================
        for f in interpolated.mean():
            feature.append(f)
        for f in interpolated.std():
            feature.append(f)
        for f in interpolated.median():
            feature.append(f)
        for f in interpolated.min():
            feature.append(f)
        for f in interpolated.max():
            feature.append(f)
        for f in interpolated.quantile(q=0.25):
            feature.append(f)

        for f in interpolated.quantile(q=0.75):
            feature.append(f)
        for f in interpolated.mode():
            feature.append(f)

        for f in interpolated.kurtosis():
            feature.append(f)

        for f in interpolated.skew():
            feature.append(f)
        # ============frequency domain=======================    
    
        for f in fft_trans.mean():
            feature.append(f)
        for f in fft_trans.std():
            feature.append(f)
        for f in fft_trans.max():
            feature.append(f)
        for f in fft_trans.min():
            feature.append(f)

        for f in fft_skew(fft_trans):
            feature.append(f)
        for f in fft_kurt(fft_trans):
            feature.append(f)

        all_features.append(feature)



    return pd.DataFrame(all_features)
```

# Extract features and omit overlap data

```python
import pandas as pd
import numpy as np

def fft_skew(fft_trans):
    colWise = []
    rowWise = []
    fft_mean, fft_std = fft_trans.mean(), fft_trans.std()
    for j in fft_trans:  # j th column among 22
        for x in fft_trans.iloc[:, j]:  # x is x value among 300

            rowWise.append(((x - fft_mean[j]) / fft_std[j]) ** 3)  # append the value for each rowwise computeation

        colWise.append(np.mean(rowWise))  # append the value for each feature
    return colWise


def fft_kurt(fft_trans):
    colWise = []
    rowWise = []
    fft_mean, fft_std = fft_trans.mean(), fft_trans.std()
    for j in fft_trans:  # j th column among 22
        for x in fft_trans.iloc[:, j]:  # x is x value among 300

            rowWise.append(((x - fft_mean[j]) / fft_std[j]) ** 4)  # append the value for each rowwise computeation

        colWise.append(np.mean(rowWise))  # append the value for each feature
    return colWise


def extraction_dp(all_intervals ):
    all_features = []
    
    for interval in all_intervals:
        interval = pd.DataFrame(interval.data)
        interval.drop(interval.index[150:], inplace=True) # omit overlap data
        #print interval
        interpolated = interval.interpolate(method='linear')
        interpolated = interpolated.fillna(method='backfill')
        fft_trans = np.abs(np.fft.fft(interpolated, axis=0))
        fft_trans = pd.DataFrame(fft_trans)
        feature = []
        # ==============time domain====================
        for f in interpolated.mean():
            feature.append(f)
        for f in interpolated.std():
            feature.append(f)
        for f in interpolated.median():
            feature.append(f)
        for f in interpolated.min():
            feature.append(f)
        for f in interpolated.max():
            feature.append(f)
        for f in interpolated.quantile(q=0.25):
            feature.append(f)

        for f in interpolated.quantile(q=0.75):
            feature.append(f)
        for f in interpolated.mode():
            feature.append(f)

        for f in interpolated.kurtosis():
            feature.append(f)

        for f in interpolated.skew():
            feature.append(f)
        # ============frequency domain=======================    
    
        for f in fft_trans.mean():
            feature.append(f)
        for f in fft_trans.std():
            feature.append(f)
        for f in fft_trans.max():
            feature.append(f)
        for f in fft_trans.min():
            feature.append(f)

        for f in fft_skew(fft_trans):
            feature.append(f)
        for f in fft_kurt(fft_trans):
            feature.append(f)

        all_features.append(feature)



    return pd.DataFrame(all_features)
```

### split data into 70/30 training set and validation set for stepwise selection of data

```python
train, test = train_test_split(data_full, test_size=0.3)
X_train =  train.loc[:, train.columns!="terrain"]
Y_train =  train.loc[:, "terrain"]

X_test =  test.loc[:, test.columns!="terrain"]
Y_test =  test.loc[:, "terrain"]
```

### perform stepwise selection of important features 

```python
mcs_score = []
featureTest = range(8, len(indices))
modelRF1 = RandomForestClassifier(max_features = 8, n_estimators=200)

Y_mul_test = label_binarize(Y_test, classes=[0, 1, 2, 3,4])
for index in featureTest:
    feature_select = features[indices[:index]]
    X_train_select = X_train[feature_select]

    #Fitting the RandomForest model
    # RandomForest with the default number of trees, which is 10. For around 30 features this is too few. 

    modelRF1.fit(X_train_select,Y_train)
    X_test_select = X_test[feature_select]
    #Predict on test set
    Y_pred1 = modelRF1.predict_proba(X_test_select)
    mcs_score.append(log_loss(Y_mul_test, Y_pred1))
```

### plot  multiclass cross entropy scores

```python
plt.figure(figsize=(20, 10))
 
plt.scatter(featureTest,mcs_score,color='blue') 

plt.xlabel('The nuber of features')
plt.ylabel('Multiclass cross entropy')

plt.savefig('D:\\kaggle machine learning\\MLproject2\\featureselection.png', dpi=100)
plt.show()
```

```python
# make it more in details
plt.figure(figsize=(20, 20))
 
plt.scatter(featureTest[:75],mcs_score[:75],color='blue') 

plt.xlabel('The nuber of features')
plt.ylabel('Multiclass cross entropy')
plt.show()
```
